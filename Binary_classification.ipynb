```python
import pandas
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
```

```python
dataframe = pandas.read_csv("/content/drive/MyDrive/PIIS/sonar.csv", header=None)
dataset = dataframe.values
X = dataset[:,0:60].astype(float)
Y = dataset[:,60]
```

```python
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)
```

```python
model = Sequential()
model.add(Dense(60, input_dim=60, kernel_initializer='normal', activation='relu'))
model.add(Dense(1, kernel_initializer='normal', activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
```

```python
net = model.fit(X, encoded_Y, epochs=100, batch_size=10, validation_split=0.1)
```

```python
from matplotlib import pyplot as plt

import numpy as np
N = np.arange(0, len(net.history["loss"]))
plt.figure()
plt.plot(N, net.history["loss"], label="train_loss")
plt.plot(N, net.history["val_loss"], label="val_loss")
plt.title("Loss")
plt.legend()


plt.figure()
plt.plot(N, net.history["accuracy"], label="train_acc")
plt.plot(N, net.history["val_accuracy"], label="val_acc")
plt.title("Accuracy")

plt.legend()

plt.show()
```

```python
model2 = Sequential()
model2.add(Dense(30, input_dim=60, kernel_initializer='normal', activation='relu'))
model2.add(Dense(1, kernel_initializer='normal', activation='sigmoid'))
```

```python
model2.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
```

```python
net2 = model2.fit(X, encoded_Y, epochs=100, batch_size=10, validation_split=0.1)
```

```python
N = np.arange(0, len(net.history["loss"]))
plt.figure()
plt.plot(N, net2.history["loss"], label="train_loss")
plt.plot(N, net2.history["val_loss"], label="val_loss")
plt.title("Loss")
plt.legend()


plt.figure()
plt.plot(N, net2.history["accuracy"], label="train_acc")
plt.plot(N, net2.history["val_accuracy"], label="val_acc")
plt.title("Accuracy")

plt.legend()

plt.show()
```

```python
score = model.evaluate(X, encoded_Y, verbose=0)
score2 = model2.evaluate(X, encoded_Y, verbose=0)
print("Loss in 1st model: {:.2f} Accuracy in 1st model: {:.2f}".format(score[0], score[1]))
print("Loss in 2nd model: {:.2f} Accuracy in 2nd model: {:.2f}".format(score2[0], score2[1]))
```
